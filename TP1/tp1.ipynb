```python
import comnumfip 
import matplotlib.pyplot as plt
def showSignal(mod,point):
    plt.plot(mod,point)

    # Des légendes, c'est toujours bien !
    plt.xlabel("t")
    plt.ylabel("w(t)")
    plt.title("Modulation")

    # Affichage (pas toujours nécessaire, mais parfois c'est indispensable)
    plt.show()

N=10
seq = comnumfip.randmary(N,[0.3, 0.7])
print(seq)
seq16 = comnumfip.bin2mary(seq,N)

modA,pointA = comnumfip.mod_a(seq,1,1)
modB,pointB = comnumfip.mod_b(seq,1,1)
modC,pointC = comnumfip.mod_c(seq,1,1)
modD,pointD = comnumfip.mod_d(seq,1,1)

modE,pointE = comnumfip.mod_e(seq16,1,1)

showSignal(modA,pointA)
showSignal(modB,pointB)
showSignal(modC,pointC)
showSignal(modD,pointD)
showSignal(modE,pointE)
```
