```python
import comnumfip 
N=10
seq = comnumfip.randmary(N,[0.3, 0.7])
print(seq)
seq16 = comnumfip.bin2mary(seq,N)
```

```python

import matplotlib.pyplot as plt
import scipy.signal as signal

def showSignal(mod,point,title):
    plt.plot(mod,point)
    # Des légendes, c'est toujours bien !
    plt.title(title)
    # Affichage (pas toujours nécessaire, mais parfois c'est indispensable)
    plt.show()



modA,pointA = comnumfip.mod_a(seq,1,1)
modB,pointB = comnumfip.mod_b(seq,1,1)
modC,pointC = comnumfip.mod_c(seq,1,1)
modD,pointD = comnumfip.mod_d(seq,1,1)

modE,pointE = comnumfip.mod_e(seq16,1,1)

showSignal(modA,pointA,"Modulation A")
showSignal(modB,pointB,"Modulation B")
showSignal(modC,pointC,"Modulation C")
showSignal(modD,pointD,"Modulation D")
showSignal(modE,pointE,"Modulation E")

#1 AMI BD
#2 BPSK Porteuse
#3 Manchester BD
#4 NRZ BD
#5 16-QAM Porteuse

xA,yA = signal.periodogram(pointA,fs=100)
xB,yB = signal.periodogram(pointB,fs=100)
xC,yC = signal.periodogram(pointC,fs=100)
xD,yD = signal.periodogram(pointD,fs=100)
xE,yE = signal.periodogram(pointE,fs=100)
showSignal(xA,yA,"Densité spectrale A")
showSignal(xB,yB,"Densité spectrale B")
showSignal(xC,yC,"Densité spectrale C")
showSignal(xD,yD,"Densité spectrale D")
showSignal(xE,yE,"Densité spectrale E")

#Harmonique dans le spectre signifie que la largeur de bande de la modulation est grande 
#On veut une largeur de bande la plus peitite possible pour pas bouffer de la bande passante
#AMI est moins bon pour la robustesse au bruit car il passe par 0
#Est ce que si on iverse ça va marcher ? Inversion de polarité 
# BUS SPI etc on va travailler en bande de base car c'est une communication par câble, Ils utilisents nrz etc...
```
